8. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es más sencillo dejarlos dentro de la figura.

```python
x = ['Ana' , 'Luis' , 'Pedro' , 'Sofia' , 'Carmen'] 
y = [160 , 180 , 175 , 190, 170] 

plt.title('Altura de mis amigos')
plt.xlabel('amigos')
plt.ylabel('alturas')
plt.bar(x,y, color = 'green')

plt.ylim(125, max(y) +5) 

plt.show()
```

9. Basándote en el gráfico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los métodos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
barplot = plt.bar(x,y)

# Recorrer cada barra del gráfico
for bar in barplot:
    # Obtener la altura de la barra
    yval = bar.get_height()
    # Escribir el valor numérico encima de la barra
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va="bottom", ha="center")


plt.title("Altura de amigos")
plt.xlabel("Amigos")
plt.ylabel("Alturas")
plt.bar(x,y, color = "green");
```

```python
fig, ax = plt.subplots() 

barplot = ax.bar(x,y ,color = 'green')

ax.bar_label(barplot, padding= 3)
ax.set_title("Altura de amigos")
ax.set_xlabel("Amigos")
ax.set_ylabel("Alturas")

ax.set_ylim(125, max(y) +5) 

plt.show()
```

21. Vamos a pintar un histograma

```python

np.random.seed(1) # cada vez que le pida N números aleatorios, me dará los mismos

df = pd.DataFrame({"Altura" : np.random.randint(low=150, high=190, size=300)})
```

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html

```python
df.plot(bins = 5, kind='hist', rwidth=0.7, title = 'alturas' , legend= False)
```

```python
# Datos
data = df["Altura"]

# Crear el histograma y obtener los bordes de los bins
counts, bins, patches = plt.hist(data, bins=5, rwidth=0.7, edgecolor='black')

# Crear etiquetas con los intervalos exactos
bin_labels = [f"{int(bins[i])}–{int(bins[i+1])}" for i in range(len(bins)-1)]

# Poner las etiquetas centradas en cada barra
plt.xticks(
    [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)],
    bin_labels
)

# Títulos y estilo
plt.title("Alturas")
plt.xlabel("Intervalos de altura")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

22. Contornos de 3D a 2D. Escribe una función que recibiendo x,y devuelva (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

Tablero: Tanto x como y van de -3 a 3 y usaremos 256 puntos.
Usa contourf, con 8 niveles (cortes), una transparencia de 0.75 y un color map de tipo 'jet'. 
Bonus: pinta las líneas de los contornos también de negro

```python
def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256 
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n) 
X, Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8 , alpha = .75, cmap = 'jet')

c = plt.contour(X,Y,f(X,Y), 8, colors = 'black' )

plt.show()
```
