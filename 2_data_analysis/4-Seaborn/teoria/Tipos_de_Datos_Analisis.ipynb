# Tipos de Datos en Análisis de Datos

## Introducción

En el análisis de datos, comprender los tipos de datos es fundamental para seleccionar las herramientas estadísticas y gráficas adecuadas.  
Cada tipo de dato determina qué tipo de análisis podemos realizar, qué transformaciones son válidas y cómo interpretar los resultados.

Los datos se dividen, de manera general, en **dos grandes categorías**:
1. **Datos cualitativos** (o categóricos)  
2. **Datos cuantitativos** (o numéricos)

A continuación, se explican en detalle.

---

## 1. Datos Cualitativos

### Definición

Los datos cualitativos describen **características o cualidades**.  
No se expresan mediante números, sino mediante **categorías** o **etiquetas** que representan un atributo o característica.  

Estos datos nos ayudan a **clasificar** o **distinguir** elementos dentro de un conjunto.

Ejemplo general:
- Color de ojos: azul, verde, marrón, gris  
- Tipo de producto: electrónico, textil, alimenticio  

Los datos cualitativos se subdividen en dos tipos: **nominales** y **ordinales**.

---

### 1.1 Datos Nominales

**Definición:**  
Son aquellos que permiten **nombrar o clasificar** elementos en categorías **sin orden lógico o jerarquía**.  
Las categorías son **mutuamente excluyentes** (una observación pertenece a una sola categoría).

**Ejemplos:**
- Género: masculino, femenino, otro  
- Nacionalidad: argentina, chilena, peruana, mexicana  
- Tipo de mascota: perro, gato, pez, ave  

**En análisis de datos:**  
- Se usan para contar frecuencias o proporciones.  
- Se representan habitualmente con **gráficos de barras** o **gráficos de pastel**.  
- No tiene sentido calcular promedios o medianas.  

**Ejemplo práctico:**  
En un estudio sobre preferencias de color en automóviles, las categorías “rojo”, “azul” y “negro” son nominales.  
Solo importa cuántas veces aparece cada una, no existe un orden entre ellas.

---

### 1.2 Datos Ordinales

**Definición:**  
Son aquellos que, además de clasificar, **tienen un orden o jerarquía**, pero **no se puede medir la distancia exacta** entre categorías.

**Ejemplos:**
- Nivel educativo: primaria, secundaria, universidad, posgrado  
- Nivel de satisfacción: bajo, medio, alto  
- Puesto en una competencia: primero, segundo, tercero  

**En análisis de datos:**  
- Se pueden ordenar, pero **no se pueden realizar operaciones aritméticas**.  
- Se suelen representar también con **barras ordenadas** o **diagramas de frecuencias**.  

**Ejemplo práctico:**  
Si encuestamos la satisfacción del cliente y las respuestas van de “muy insatisfecho” a “muy satisfecho”, las categorías tienen un orden, pero no podemos decir que “muy satisfecho” sea el doble de “satisfecho”.

---

## 2. Datos Cuantitativos

### Definición

Los datos cuantitativos representan **cantidades numéricas**.  
Permiten realizar operaciones matemáticas como sumas, promedios, o cálculos estadísticos.  

Estos datos expresan **magnitudes medibles** y pueden dividirse en dos tipos: **discretos** y **continuos**.

---

### 2.1 Datos Discretos

**Definición:**  
Son datos numéricos que **solo pueden tomar valores enteros**.  
Se obtienen mediante **conteos**.

**Ejemplos:**
- Número de hijos por familia  
- Cantidad de llamadas recibidas por día  
- Número de productos vendidos  

**En análisis de datos:**  
- Se representan con **gráficos de barras o diagramas de dispersión con puntos enteros**.  
- No tiene sentido hablar de valores intermedios (no hay 2,5 hijos).  

**Ejemplo práctico:**  
En un call center, el número de llamadas atendidas por operador (10, 15, 22) es discreto, porque no se pueden registrar valores decimales.

---

### 2.2 Datos Continuos

**Definición:**  
Son datos numéricos que pueden tomar **cualquier valor dentro de un rango**, incluyendo fracciones o decimales.  
Se obtienen mediante **mediciones**.

**Ejemplos:**
- Peso corporal (kg)  
- Temperatura (°C)  
- Tiempo de ejecución (segundos)  
- Altura (cm)  

**En análisis de datos:**  
- Se representan con **histogramas** o **diagramas de densidad**.  
- Permiten calcular estadísticas como media, varianza, desviación estándar, etc.  

**Ejemplo práctico:**  
Si medimos la temperatura de una ciudad durante un día, los valores pueden variar continuamente (22.1°C, 22.5°C, 22.6°C…), no se limitan a números enteros.

---

## 3. Comparación General

| Tipo de dato | Subtipo | Naturaleza | Ejemplos | Tipo de análisis |
|---------------|----------|-------------|-----------|------------------|
| Cualitativo | Nominal | Categórico sin orden | Género, nacionalidad, color | Conteo, proporciones, moda |
| Cualitativo | Ordinal | Categórico con orden | Nivel educativo, satisfacción | Ordenamiento, mediana |
| Cuantitativo | Discreto | Numérico entero | Número de hijos, llamadas, autos | Frecuencias, promedios |
| Cuantitativo | Continuo | Numérico decimal | Peso, altura, temperatura | Estadísticas descriptivas, histogramas |

---

## Conclusión

Identificar correctamente el tipo de dato es un paso esencial en cualquier análisis de datos.  
De ello depende qué tipo de gráfico utilizar, qué medidas de resumen aplicar y qué técnicas estadísticas son válidas.

- Los **datos cualitativos** describen características.  
- Los **datos cuantitativos** miden magnitudes.  
- Dentro de cada grupo, distinguir entre **nominal/ordinal** y **discreto/continuo** garantiza un análisis correcto y significativo.
